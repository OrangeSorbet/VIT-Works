# Basics of Pandas

#### basic intuition

```python
a = [34, 24, 55, 67, 12, 35, 65, 77, 33, 30]
b = [45, 25, 69, 43, 10, 54, 28, 95, 18, 50]

c = [a[i] + b[i] for i in range(len(a))]
print(c)
```

#### with numpy?

```python
a = np.array([34, 24, 55, 67, 12, 35, 65, 77, 33, 30])
b = np.array([45, 25, 69, 43, 10, 54, 28, 95, 18, 50])

print(a+b)
```

```python
c = np.mean(a)
means = [a+b/2]
print(c)
print(means)
```
