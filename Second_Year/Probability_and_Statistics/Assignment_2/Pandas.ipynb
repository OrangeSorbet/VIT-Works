```python
def mean(df):
    return df.mean()

def meanmanual(df):
    for i in range(df.shape[1]):
        col_sum = 0
        for j in range(df.shape[0]):
            col_sum += df.iat[j, i]
        col_mean = col_sum / df.shape[0]
        print(f"Mean of column {df.columns[i]}: {col_mean}")
```

```python
print(mean(df))
meanmanual(df)
```

```python
def median(df):
    return df.median()

def medianmanual(df):
    for i in range(df.shape[1]):
        col = sorted([df.iat[j, i] for j in range(df.shape[0])])
        n = len(col)
        mid = n // 2
        if n & 1:
            col_median = col[mid]
        else:
            col_median = (col[mid - 1] + col[mid]) / 2
        return col_median
```

```python
print(median(df))
print(f"Median of {df.columns[1]}: {medianmanual(df)}")
```

```python
temp = np.array([2, 9, 8, 6, 4])

variance = np.mean((temp - np.mean(temp))**2)
print("Variance is -", variance)
```

```python
def quartiles(df):
    Q2 = medianmanual(df)
    Q1 = medianmanual(df[df < Q2])
    Q3 = medianmanual(df[df > Q2])
    return Q1, Q2, Q3

print(quartiles(df))
```
