```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
def nCr(n, r):
    if r > n:
        return 0
    else:
        return factorial(n) // (factorial(r) * factorial(n - r))
    
def binomial(n, r, p):
    return nCr(n, r) * (p ** r) * ((1 - p) ** (n - r))

def binomial_cdf(n, low, up, p):
    cdf = 0
    for r in range(low, up + 1):
        cdf += binomial(n, r, p)
    return cdf
```

```python
"""
A coin is tossed 7 times - 
1. To find probability of getting 4 heads
2. To find probability of getting at least 4 heads
"""
n = 7
p = 0.5
# 1. Probability of getting exactly 4 heads
r = 4
exact_probability = binomial(n, r, p)
print(f"Probability of getting exactly 4 heads: {exact_probability}")
# 2. Probability of getting at least 4 heads (4, 5, 6, or 7 heads)
cumulative_probability = binomial_cdf(n, 4, n, p)
print(f"Probability of getting at least 4 heads: {cumulative_probability}")
```
